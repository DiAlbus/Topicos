### 1. Resuelva los problemas de valor inicial usando la transformada de Laplace.

```python
from sympy import *
init_printing()
```

```python
#Se definen las variables
x, s = symbols('x s')
y = Function('y')
```

```python
#Se aplica la transformada de Laplace (Y es la tansformada de y)
Y = laplace_transform(y(x), x, s)
Y
```

---

#### (a) $ y''+4y = 0, \ \ y(0) = 2, \ \ y'(0) = -8 $

```python
#Se define la ecuación diferencial
ode = Eq( diff(y(x),x, 2) + 4*y(x), 0)
ode
```

```python
#Se definen las condiciones iniciales
ics = {y(0): 2, Subs(diff(y(x),x),x,0): -8}
ics
```

```python
#L_ode es la ecuación después de aplicar la transformada de Laplace
L_ode = Eq( laplace_transform(ode.lhs, x, s)[0].subs(ics), laplace_transform(ode.rhs, x, s)[0].subs(ics) )
L_ode
```

```python
#Se resuelve la ecuación
L_sol = solve(L_ode, Y)[0]
L_sol
```

```python
#Se aplica la transformada inversa (recordar que theta(t) es la función escalón)
sol = inverse_laplace_transform(L_sol[0], s, x)
sol
```

```python
#Se grafica la solución
c = plotting.plot(sol)
```
